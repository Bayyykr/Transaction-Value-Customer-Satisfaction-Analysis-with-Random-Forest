```python
import pandas as pd

data = pd.read_csv('dataset_performa_penjualan.csv')
data
```

```python
data.isna().sum()
```

```python
data['tanggal'] = pd.to_datetime(data['tanggal'], infer_datetime_format=True)
data['tanggal'] = data['tanggal'].dt.strftime('%d-%m-%Y')
```

```python
data['kategori'] = data['kategori'].astype(str).str.lower()
data
```

```python
data[['jumlah', 'rating']] = data[['jumlah', 'rating']].fillna(data[['jumlah', 'rating']].mean())
data.isna().sum()
```

```python
data['harga'] = data.groupby('kategori')['harga'].transform(lambda x: x.fillna(x.mean()))
data
```

```python
import numpy as np

data[ 'jumlah'] = np.ceil(data['jumlah']).astype(int)
data
```

```python
data['harga'] = data['harga'].astype(int)
data
```

```python
data['komentar']=data['komentar'].fillna('')
```

```python
data.groupby('kota')['id_transaksi'].sum()
```

```python
data['kota'].unique()
```

```python
data['kota'] = data['kota'].str.strip().str.title()
```

```python
data.groupby('kota')['id_transaksi'].sum()
```

```python
def remove_outlier_per_kategori(df):
    Q1 = df['harga'].quantile(0.25)
    Q3 = df['harga'].quantile(0.75)
    IQR = Q3 - Q1
    low = Q1 - 1.5 * IQR
    high = Q3 + 1.5 * IQR
    return df[(df['harga'] >= low) & (df['harga'] <= high)]

data_clean = data.groupby('kategori', group_keys=False).apply(remove_outlier_per_kategori)
print(len(data))
print(len(data_clean))
```
