```python
import pandas as pd

data = pd.read_csv('data_clean.csv')
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
columns = ['kategori', 'kota']
for column in columns:
    data[column] = le.fit_transform(data[column])
data
```

```python
data['total_transaksi'] = data['harga'] * data['jumlah']
data.head()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

X = data[['kategori', 'kota', 'weekday_or_weekend', 'day', 'month', 'year', 'harga', 'jumlah']]
y = np.log1p(data['total_transaksi'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f'X_train-Shape: {X_train.shape}')
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

parameters= {
    'n_estimators': [100, 200, 300, 500],
    'max_depth': [None, 10, 20, 30, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_grid=parameters,
    n_jobs = 1,
    verbose = 1,
    scoring='r2'
)

grid_search.fit(X_train, y_train)
```

```python
print(f'Best Score: {grid_search.best_score_}')
```

```python
best_params = grid_search.best_estimator_.get_params()
print('Best Parameters:')
for param in parameters:
    print(f'\t{param}: {best_params[param]}')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred_log = grid_search.predict(X_test)
y_pred = np.expm1(y_pred_log)
y_test_real = np.expm1(y_test)
r_square = r2_score(y_test_real, y_pred)
MAE = mean_absolute_error(y_test_real, y_pred)
MSE = mean_squared_error(y_test_real, y_pred)

print(f'R-Square: {r_square}')
print(f'MAE: {MAE}')
print(f'MSE: {MSE}')
```

```python
data[['harga','jumlah','total_transaksi']].describe()
```

```python
best_model = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)
print("Best CV Score:", grid_search.best_score_)
```

```python
importances = best_model.feature_importances_
importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print("\nFeature Importances:")
print(importance_df)
test_score = best_model.score(X_test, y_test)
print("\nTest Accuracy:", test_score)
```
